```python
import exantedata_api as ed
```

```python
# 1. Obtaining a token
token = ed.getToken(
    username='username',
    password='password'
)
```

```python
# 2. Specifying a ticker
tickerQuery = 'TR.FOREIGNFX.SHORT.NET.D'
```

```python
# 3.1 Metadata query
metadata = ed.getMetaData(
    token=token,
    tickerQuery=tickerQuery
)

metadata
```

```python
# 3.2 Data queries

# Basic usage
data = ed.getData(
    token=token,
    tickerQuery=tickerQuery
)

data
```

```python
# startDate and endDate params usage
data = ed.getData(
    token=token,
    tickerQuery=tickerQuery,
    startDate='2023-01-01',
    endDate='2023-12-31',
)

data
```

```python
# pediod param usage
data = ed.getData(
    token=token,
    tickerQuery='TR.FOREIGNFX.SHORT.NET.M',
    startDate='2023-01-01',
    endDate='2023-12-31',
    period='begin'
)

data
```

```python
# freq param usage
data = ed.getData(
    token=token,
    tickerQuery='TR.FOREIGNFX.SHORT.NET.D',
    startDate='2023-10-01',
    endDate='2023-12-31',
    freq='W'
)

data
```

```python
# agg_method param usage
data = ed.getData(
    token=token,
    tickerQuery='TR.FOREIGNFX.SHORT.NET.D',
    startDate='2023-10-01',
    endDate='2023-12-31',
    freq='W',
    agg_method='sum',
)

data
```

```python
# fill_method param usage
data = ed.getData(
    token=token,
    tickerQuery='TR.FOREIGNFX.SHORT.NET.M, TR.HFPROXFLOW.EQ.IN.W',
    startDate='2023-01-01',
    endDate='2023-12-31',
    fill_method='ffill',
)

data
```

```python
# fill_value param usage
data = ed.getData(
    token=token,
    tickerQuery='TR.FOREIGNFX.SHORT.NET.M, TR.HFPROXFLOW.EQ.IN.W',
    fill_value=0,
)

data
```

```python
# 4. Specifying proxies

# Obtaining a token with proxies
token = ed.getToken(
    username='username',
    password='password',
    proxies={
        'http': 'http://<IPADDR>:<PORT>',
        'https': 'https://<IPADDR>:<PORT>'
    }
)
```

```python
# Metadata query with proxies
metadata = ed.queryData(
    token=token,
    tickerQuery=tickerQuery,
    proxies={
        'http': 'http://<IPADDR>:<PORT>',
        'https': 'https://<IPADDR>:<PORT>'
    }
)
```

```python
# Data query with proxies
data = ed.queryData(
    token=token,
    tickerQuery=tickerQuery,
    proxies={
        'http': 'http://<IPADDR>:<PORT>',
        'https': 'https://<IPADDR>:<PORT>'
    }
)
```
